```python
# Import our dependencies
import pandas as pd
import os
from pathlib import Path
import csv
import matplotlib.pyplot as plt
import numpy as np
```

```python
#read the inpout file
input_file = input("Please enter the path and name of the file to be analyzed ? :")

file_check =  Path(input_file)
file_exists = file_check.exists()
print(file_exists)
```

```python
file_one_df = pd.read_excel(input_file, encoding = "ISO-8859-1")
file_one_df.head()

# get the data size
data_set_size = file_one_df.shape
print(data_set_size)

#get all column names to create the next data sets
file_one_df.columns.values
```

```python
## Select the columns you want to analyze
df1=pd.DataFrame()

df1 = file_one_df.filter(['GSS YEAR FOR THIS RESPONDENT                       ','LABOR FORCE STATUS_labels','R SELF-EMP OR WORKS FOR SOMEBODY_labels'
                          ,'MARITAL STATUS_labels','TRAVEL TIME TO WORK_labels','TOTAL FAMILY INCOME_labels','POLITICAL PARTY AFFILIATION_labels',
                         "R'S HH POVERTY STATUS_labels",'DISTANCE BELOW POVERTY LINE_labels','GENERAL HAPPINESS_labels',
                         'HAPPINESS OF MARRIAGE_labels','CONDITION OF HEALTH_labels'],axis=1)
df1.head()
```

```python
#lets rename teh columns
df1 = df1.rename(columns={'GSS YEAR FOR THIS RESPONDENT                       ': 'Year', 
                          'LABOR FORCE STATUS_labels': 'Labor_Force_Status',
                         'R SELF-EMP OR WORKS FOR SOMEBODY_labels' : 'Self_emp_status',
                         'MARITAL STATUS_labels' : 'Marital_status',
                         'TRAVEL TIME TO WORK_labels': "Travel_time_to_work",
                         'TOTAL FAMILY INCOME_labels' : 'Total_Family_income',
                         'POLITICAL PARTY AFFILIATION_labels' : 'Political_affiliation',
                         'R''S HH POVERTY STATUS_labels':'Household_povery_label',
                         'DISTANCE BELOW POVERTY LINE_labels' :'Distance_Below_Poverty_Line',
                         'GENERAL HAPPINESS_labels' : 'General_happiness',
                         'HAPPINESS OF MARRIAGE_labels' : 'Happiness_of_marraige',
                         'CONDITION OF HEALTH_labels' : 'Condition_of_health'})

df1.head
```

```python
#I think we can use this data to show Economic Wellness
happytimes_pd = df1.groupby(["General_happiness" ,"Total_Family_income" ,'Condition_of_health']).count()
happytimes_pd

#need to work how to interpret this data  for visulization
```
